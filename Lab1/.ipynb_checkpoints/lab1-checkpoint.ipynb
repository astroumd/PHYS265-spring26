## Part 1: The Gravitational Potential of the Earth
----

The gravitational potential at a distance $r$ from a mass $M$ is given by:

$$\Phi(r) = -\frac{GM}{r}$$

where G is the gravitational constant.

1. Code up a function that takes as an input the mass of a body $M$ , the cartesian location of the
mass $(x_M , y_M)$, and the location of some distant evaluation point $(x, y)$. Have the function
return the potential at the evaluation point. You may consider approximating the Earth as
a point particle, but in which case, be careful! For in that approximation, the the potential
contains a singularity. In our function, use all modern best practices for coding.
E.g. type hints, type checks, pre and post conditions, doc strings, etc.

```python
def grav_potential (mass, loc_e, loc_m=[0.0,0.0]):
    """Vectorizable function that calculates the gravitational potential at a distance from a mass M
    Args: 
        mass (float): mass of body M
        loc_m (list): location of the mass with length of 2 (default at 0.0,0.0)
        loc_e (list): loction of evaluation with length of 2

    Returns:
        float: gravitational potential
    """
    assert isinstance(mass, float) and mass > 0, "Mass should be a positive, non-zero float"
    assert isinstance(loc_e, list) and len(loc_e) == 2, "Check that the location of evaluation is a list of 2"
    assert isinstance(loc_m, list) and len(loc_m) == 2, "Check that the location of the mass is a list of 2"
    loc_e = np.array(loc_e)
    loc_m = np.array(loc_m)
    assert np.issubdtype(loc_e.dtype, np.floating), "Location of evaluation should be a float or array of floats"
    assert np.issubdtype(loc_m.dtype, np.floating), "Location of the mass should be a float or array of floats"
    r = np.hypot(loc_e[0]-loc_m[0], loc_e[1]-loc_m[1])
    potential = np.zeros(np.shape(r))
    mask = np.isclose(r, 0)
    _mask = np.logical_not(mask)
    potential[mask] = np.nan
    potential[_mask] = -G*mass/r[_mask]
    assert np.issubdtype(potential.dtype, np.floating) and (potential<0).any(), "Output should be an array of negative floats"
    return potential
```

2. Make a 1D plot of the absolute value of the gravitational potential $|\Phi|$ as a function of distance $x$ from the surface of the Earth, assuming $y = 0$.
Plot out to $1.5\times$ the distance from the Earth to the Moon. Because the potential falls quickly with r, make the y-axis logarithmic to visualize it better.

```python
x = np.linspace(0, 1.5*dem, 50)
y = grav_potential(mass_earth, [x+rad_earth, np.zeros(len(x))])

fig, ax = plt.subplots()
ax.plot(x, np.abs(y))
ax.set_yscale('log')
ax.set_title("Gravitational Potential vs. Distance from Earth's Surface")
ax.set_xlabel("Distance from Earth [m]")
ax.set_ylabel("log|Gravitational Potential| [V]")
```

## Part 3: The Gravitational Force Field of the Earth-Moon System
----
The gravitational force $\vec{F}$ 
that a mass $M_1$ exerts on a mass $m_2$ is:

$$\vec{F}_{21} = -G \frac{M_1 m_2}{|\vec{r}_{21}|^2} \hat{r}_{21}$$

where $\vec{r}_{21}$ is the displacement vector from $M_1$ to $m_2$.

1. Code up a function that takes as an input the mass of a body $M_1$, 
the mass of a secondary body $m_2$,
the cartesian location of the first mass $(x_1, y_1)$, 
and the cartesian location of the second body $(x_2, y_2)$. 
Have the function return the two gravitational force components, 
$F_x$ and $F_y$, experienced by the mass $m_2$ 
from the force exerted by $M_1$.
Note that the gravitational force contains a singularity, 
and so if you made the approximation that the Earth is a finite point,
you will need to avoid it.
Like for the 1D potential, make sure it follows modern best-coding practices.

```python
def grav_force(mass_1, mass_2, loc_1, loc_2):
    """Vectorizable function that calculates the gravitational force between two bodies
    Args: 
        mass_1 (float): mass of the primary body
        mass_2 (float): mass of the secondary body
        loc_1 (list): location of the mass of the primary body with a length of 2
        loc_1 (list): loction of the mass of the secondary body with length of 2

    Returns:
        float: gravitational force component x
        float: gravitational force component y
    """
    assert isinstance(mass_1, float) and mass_1 > 0, "Primary mass should be a positive, non-zero float"
    assert isinstance(mass_2, float) and mass_2 > 0, "Secondary mass should be a positive, non-zero float"
    assert isinstance(loc_1, list) and len(loc_1) == 2, "Check that the location of the primary body is a list of 2"
    assert isinstance(loc_2, list) and len(loc_2) == 2, "Check that the location of the secondary body is a list of 2"
    loc_1 = np.array(loc_1)
    loc_2 = np.array(loc_2)
    assert np.issubdtype(loc_1.dtype, np.floating), "Location of the primary body should be a float or array of floats"
    assert np.issubdtype(loc_2.dtype, np.floating), "Location of the secondary body should be a float or array of floats"
    r = np.hypot(loc_2[0]-loc_1[0], loc_2[1]-loc_1[1])
    vec_r = np.array([loc_2[0]-loc_1[0], loc_2[1]-loc_1[1]])
    vec_r_hat = vec_r/np.linalg.norm(vec_r)
    force_x = np.zeros(np.shape(r))
    force_y = np.zeros(np.shape(r))
    mask = np.isclose(r, 0)
    _mask = np.logical_not(mask)
    force_x[mask] = np.nan
    force_y[mask] = np.nan
    force_x[_mask] = -G*mass_1*mass_2/(r[_mask]**2)*vec_r_hat[0][_mask]
    force_y[_mask] = -G*mass_1*mass_2/(r[_mask]**2)*vec_r_hat[1][_mask]
    assert np.issubdtype(force_x.dtype, np.floating), "Output for x component should be an array of floats"
    assert np.issubdtype(force_y.dtype, np.floating), "Output for y component should be an array of floats"
    return force_x, force_y
```

3. You will note there is a point between the Earth and the Moon where the field lines seem to vanish. This is the point where the Earth's and Moon's gravitational forces cancel eachother. By setting $|\vec{F}_{moon}| = |\vec{F}_{earth}|$, calculate the radial distance of the "balance point." From there, identify the $x$ and $y$ coordinates of the balance point. Typeset your derivation (show multiple steps!) in latex in a markdown cell below. Then, calculate it numerically, and draw the answer onto a new version of your force field plot. Again mark the location of the earth with a black circle, and the moon with a black square; mark the location of the balance point with a red x. Add a legend.

**Your Derivation Goes in the markdown cell below:**

$|\vec{F}_{moon}| = |\vec{F}_{earth}|$  
$G\frac{M_{moon} m}{r_{moon}^2} = G\frac{M_{earth} m}{r_{earth}^2}$  
$\frac{M_{moon}}{r_{moon}^2} = \frac{M_{earth}}{r_{earth}^2}$  
$r_{moon}^2 = r_{earth}^2\frac{M_{moon}}{M_{earth}}$  
$r_{moon} = r_{earth}\sqrt{\frac{M_{moon}}{M_{earth}}}$  
Given: $d_{EM} = r_{earth} + r_{moon}$ along the path E-M  
$d_{EM} = r_{earth} + r_{earth}\sqrt{\frac{M_{moon}}{M_{earth}}}$  
$d_{EM} = r_{earth}(1 + \sqrt{\frac{M_{moon}}{M_{earth}}})$    
$r_{earth} = \frac{d_{EM}}{1 + \sqrt{M_{moon}/M_{earth}}}$  
As a vector along path E-M:$$\vec{r}_{earth} = \frac{d_{EM}}{1 + \sqrt{M_{moon}/M_{earth}}}\hat{r}_{EM}$$

And your updated plot, including the balance point, goes in the cell below.

```python
vec_r = np.array([dem/np.sqrt(2), dem/np.sqrt(2)])
vec_r_hat = vec_r/np.linalg.norm(vec_r)
r_mag = dem/(1 + np.sqrt(mass_moon/mass_earth))
bp_x = r_mag * vec_r_hat[0]
bp_y = r_mag * vec_r_hat[1]
print(f"Balance point distance from the moon: {dem-r_mag:.2e} m")

fig, ax = plt.subplots()
stream = ax.streamplot(x_grid, y_grid, Fx, Fy, color=colors, cmap=cm.plasma)
cbar = fig.colorbar(stream.lines, label="log|Gravitational Force| [N]")
ax.set_xlabel("Distance from Earth [m]")
ax.set_ylabel("Distance from Earth [m]")
ax.set_title("Gravitational Force of the Earth-Moon System on Apollo 11")
ax.plot(0, 0, "ok", label="Earth", ms=10)
ax.plot(dem/np.sqrt(2), dem/np.sqrt(2), "sk", label="Moon", ms=8)
ax.plot(bp_x, bp_y, "xr", label="Balance point", ms=8)
ax.legend(loc="upper left")
```

## Part 4: Altitude of the Saturn V Rocket
----
Rockets function by conservation of momentum.
The ejection of fuel backwards propels the rocket forwards.
The change in a rocket's velocity $\Delta v$ as a 
function of time $t$ (under a set of simplifying assumptions) 
is given by the Tsiolkovsky rocket equation:

$$\Delta v(t) = v_e \ln \left(\frac{m_0}{m(t)}\right) - gt$$

where $m_0$ is the initial "wet" mass (fuel + rocket parts + payload), 
$m(t) = m_0 - \dot{m}t$ is the mass at time $t$, 
$\dot{m}$ is the fuel burn rate (which we will assume to be constant), 
$v_e$ is the fuel exhaust velocity, and $g$ is the gravitational acceleration.

You can find the altitude $h$ of the rocket at "burnout" 
(when all the fuel is used up) by integrating the velocity over time:

$$h = \int_{0}^{T} \Delta v(t) \, dt$$

where $T$ is the total burn time of the rocket, 
which by conservation of momentum arguments is:

$$T = \frac{m_0 - m_f}{\dot{m}}$$

where $m_f$ is the final "dry mass" of the rocket, 
once all the fuel is burned.

1. Calculate the burn time $T$ for the first stage of
the new Saturn V rocket that NASA is planning
to use to carry the Apollo capsule.

```python
T = (m0 - mf)/mr
print(f"T = {T:.0f}s")
```

2. Define a function that returns the change in velocity $\Delta v(t)$.
It should take all necessary arguments, meaning: the time, 
wet mass, dry mass, burn rate, exhaust speed, and gravitational constant $g$. 
Have it return the change in velocity.
Note that the rocket eventually runs out of fuel, 
so you should design your function carefully so that 
$\Delta v$ eventually becomes zero after all fuel is spent.
Again, document and guard your function according to best practices.

```python
def delta_v(t, m0=m0, mf=mf, mr=mr, ve=ve, g=g): 
    """Vectorizable function that calculates the change in velocity of a missile
    Args: 
        t (float): time
        m0 (float): wet mass (default to global constants)
        mf (float): dry mass (default to global constants)
        mr (float): burn rate (default to global constants)
        ve (float): exhaust speed (default to global constants)
        g (float): gravitational constant(default to global constants)

    Returns:
        float: change in velocity
    """
    t = np.array(t)
    assert np.issubdtype(t.dtype, np.floating), "Time should be a float or array of floats"
    assert isinstance(m0, float) and m0>0, "Wet mass should be a positive, non-zero float"
    assert isinstance(mf, float) and mf>0, "Dry mass should be a positive, non-zero float"
    assert isinstance(mr, float) and m0>0, "Burn rate should be a positive, non-zero float"
    assert isinstance(ve, float) and m0>0, "Exhaust speed should be a positive, non-zero float"
    assert isinstance(g, float) and g>0, "Gravitational constant should be a positive, non-zero float"
    m_func = m0 - mr*t
    change = np.zeros(np.shape(t))
    change[m_func<=mf] = 0.0 
    change[m_func>mf] = ve*np.log(m0/m_func[m_func>0]) - g*t[m_func>0]
    assert np.issubdtype(t.dtype, np.floating), "Output should be a float or array of floats"
    return change
```

3. Compute the altitude of the rocket at the end of the burn.
Do this by numerically integrating from $t=0$ (launch) to $t=T$ (burnout).
Use the `scipy.integrate.quad()` function.

```python
h, err = quad(delta_v, 0, T)
print(f"h = {h/1000:.0f}km")
```
